# Network Analysis

```python
import networkx as nx
import networkx.algorithms.shortest_paths.generic as nxpath
```

```python
hero_network = nx.from_pandas_edgelist(marvel_hero_network, source = "hero1", target = "hero2")
```

```python
hero_graph = nx.Graph(hero_network)
```

```python
plt.figure(figsize = (20, 20))
plt.style.use("fivethirtyeight")
pos = nx.spring_layout(hero_graph, k = .1)
nx.draw_networkx(hero_graph, pos, node_size = 35, node_color = "navy", edge_color = "grey", with_labels = True)
plt.show()
```

```python
graph = nx.Graph()
```

```python
# Populate the network with dataframe

# add edges
for index, row in marvel_hero_network.iterrows():
    graph.add_edge(row[0], row[1], weight = 1)

# add nodes
graph.add_nodes_from(nodes)
```

```python
# Create a list of networks for all the superheroes
# list = [graph]
```

```python
print(nx.info(graph))
```

```python
print(nx.info(hero_graph))
```

```python
# General Connection
nx.degree(graph)
```

```python
# 1. Most Influential - Degree Centrality
most_influential_hero = nx.degree_centrality(graph)
count = 0

for index in sorted(most_influential_hero, key = most_influential_hero.get, reverse = True):
    count = count + 1
    if count == 10:
        break
    print(index, '{:.2f}'.format(most_influential_hero[index]))
```

```python
# from Degree Centrality result above, we can see that Spider-Man / Peter Parker is the most influential hero
```

```python
heroes = []
degree_centrality = []
for key, score in most_influential_hero.items():
    heroes.append(key)
    degree_centrality.append(score)
```

```python
most_influential_hero = pd.DataFrame() # create a DataFrame for hero network degree centrality
most_influential_hero["Marvel Superheroes"] = heroes # I put heroes into a column that I create and name "Marvel Superheroes"
most_influential_hero["Degree Centrality"] = degree_centrality # I put degree centrality value / score into a column that I create and name "Degree Centrality"
most_influential_hero = most_influential_hero.sort_values(by = "Degree Centrality", ascending = False) # sort dataframe in descending order
```

```python
most_influential_hero
```

```python
# VISUALIZATION
plt.figure(figsize = (30, 30))
sb.barplot(x = "Degree Centrality", y = "Marvel Superheroes", data = most_influential_hero[:10])
plt.xlabel("Degree Centrality")
plt.ylabel("Marvel Superheroes")
plt.title("Top 10 Most Influential Marvel Superheroes in Marvel Universe Network")
```

```python
plt.figure(figsize = (30, 30))
sb.relplot(x = "Degree Centrality", y = "Marvel Superheroes", data = most_influential_hero[:10])
plt.xlabel("Degree Centrality")
plt.ylabel("Marvel Superheroes")
plt.title("Top 10 Most Influential Marvel Superheroes in Marvel Universe Network")
```

Conclusion: From the barplot and relplot above, we can conclude that Captain America is the most influential superhero in Marvel Universe Network

```python
most_important_connection = nx.eigenvector_centrality(hero_network)
count = 0

for index in sorted(most_important_connection, key = most_important_connection.get, reverse = True):
    count = count + 1
    if count == 10:
        break
    print(index, '{:.2f}'.format(most_important_connection[index]))
```

```python
heroes2 = []
ev_centrality = []
for key, score in most_important_connection.items():
    heroes2.append(key)
    ev_centrality.append(score)
```

```python
most_important_connection = pd.DataFrame()
most_important_connection["Marvel Superheroes"] = heroes2
most_important_connection["Eigenvector Centrality"] = ev_centrality
most_important_connection = most_important_connection.sort_values(by = "Eigenvector Centrality", ascending = False)
```

```python
most_important_connection
```

```python
plt.figure(figsize = (30, 30))
sb.barplot(x = "Eigenvector Centrality", y = "Marvel Superheroes", data = most_important_connection[:10])
plt.xlabel("Eigenvector Centrality Score")
plt.ylabel("Marvel Superheroes")
plt.title("Top 10 Most Important Connection of Marvel Superheroes in Marvel Universe Network")
```

```python
plt.figure(figsize = (30, 30))
sb.boxplot(x = "Eigenvector Centrality", y = "Marvel Superheroes", data = most_important_connection[:10])
plt.xlabel("Eigenvector Centrality Score")
plt.ylabel("Marvel Superheroes")
plt.title("Top 10 Most Important Connection of Marvel Superheroes in Marvel Universe Network")
```

From the barplot and boxplot above, we can conclude that Captain America has most important connections

```python
best_connector = nx.betweenness_centrality(hero_network)
count = 0

for index in sorted(best_connector, key = best_connector.get, reverse = True):
    count = count + 1
    if count == 10:
        break
    print(index, "{:.2f}".format(best_connector[index]))
```

```python
heroes3 = []
betweenness_centrality = []
for key, score in best_connector.items():
    heroes3.append(key)
    betweenness_centrality.append(score)
```

```python
best_connector = pd.DataFrame()
best_connector["Marvel Superheroes"] = heroes3
best_connector["Betweenness Centrality"] = betweenness_centrality
best_connector = best_connector.sort_values(by = "Betweenness Centrality", ascending = False)
```

```python
best_connector
```

```python
plt.figure(figsize = (30, 30))
sb.barplot(x = "Marvel Superheroes", y = "Betweenness Centrality", data = best_connector[:10])
plt.xlabel("Marvel Superheroes")
plt.ylabel("Betweenness Centrality")
plt.title("Top 10 Best Connector Marvel Superheroes in Marvel Universe Network")
```

```python
plt.figure(figsize = (30, 30))
sb.boxplot(x = "Marvel Superheroes", y = "Betweenness Centrality", data = best_connector[:10])
plt.xlabel("Marvel Superheroes")
plt.ylabel("Betweenness Centrality")
plt.title("Top 10 Best Connector Marvel Superheroes in Marvel Universe Network")
```

From the barplot and boxplot above, with the help of https://en.wikipedia.org/wiki/Betweenness_centrality, I can conclude that Spider-Man / Peter Parker is the best connector, which means he has the highest control over the network since the biggest amount of information passes through him.

### Algebraic Connectivity / Fidler Value  
  
  
- 1. The Algebraic Connectivity is equal to 0 if and only if the graph is disconnected.
- 2. The Algebraic Connectivity of a graph which isn't greater than the node connnectivity means the network is robust

```python
nx.node_connectivity(graph)
```

```python
nx.algebraic_connectivity(graph)
```

```python
nx.node_connectivity(hero_graph)
```

```python
nx.algebraic_connectivity(hero_graph)
```

Since algebraic connectivity / fidler value and node connectivity is 0, I can conclude that the graph / network hasn't been connected yet.
